```python
#import essential libraries

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
columns = ['Sepal length', 'Sepal width', 'Petal length', 'Petal width', 'Class_labels']
#loading data
df = pd.read_csv('bezdekIris.data', names=columns)
df.head()
```

```python
df.describe()
```

```python
#Visualizing the dataset
sns.pairplot(df, hue='Class_labels')
```

```python
#Separating features and target
data= df.values

X=data[:,0:4]
Y=data[:,4]
print(X)
print(Y)
```

```python
#Splitting the data to train and test dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
print(X_test)
```

```python
#Model 1:Support vector machine algorithm
from sklearn.svm import SVC

model_svc = SVC()
model_svc.fit(X_train, y_train)
```

```python
predic1 = model_svc.predict(X_test)
#Calculating accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, predic1)*100)
for i in range(len(predic1)):
  print(y_test[i],predic1[i])
```

```python
#Model2:Logistic Regression
from sklearn.linear_model import LogisticRegression
model_LR=LogisticRegression()
model_LR.fit(X_train,y_train)
```

```python
predic2 = model_LR.predict(X_test)
#Calculating accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, predic2)*100)
```

```python
#Model3:DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
model_DTC=DecisionTreeClassifier()
model_DTC.fit(X_train,y_train)
```
