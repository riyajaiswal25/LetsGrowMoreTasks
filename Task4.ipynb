```python
#importing libraries
import cv2
```

```python
#Reading image
image = cv2.imread("img.png")
```

```python
#reading image in rgb format
import matplotlib.pyplot as plt
%matplotlib inline
img_rgb=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
#Create new image by converting original image to greyscale
grey_image=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
grey_image
```

```python
plt.imshow(grey_image)
```

```python
#Inverting image
inv = 255 - grey_image
plt.imshow(inv)
```

```python
#Blur image by using the Gaussian Function
blur = cv2.GaussianBlur(inv, (21, 21), 0)
plt.imshow(blur)
```

```python
#Invert the blurred image for getting final,i.e, pencil sketch
invblur=255-blur
pen=cv2.divide(grey_image, invblur, scale=256.0)
plt.imshow(pen)
```
